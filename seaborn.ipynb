# DISTRIBUTION PLOTS

### distribution plot

to see the distribution of data

    we can see the ditribution of data for a single variable.
    we get a histogram

```python
sns.distplot(crash_df['not_distracted'],kde=True,bins=25)
```

kde- kernal density estimation 

bins- no of units the data is divided. here the data is divided in between 0-25 as 25 units

### joint plot

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='regression')
```

the plot is a regression plot, along with the distribution plots for the speeding and alcohol.

Note: here we have directly passed x,y as speeding and alcohol respectively, instead of crash_df[speeding] ... like that, see here we are passing another feature also, that is data=crash_df

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='kde')
```

## bar plot

## box plot

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

#### here, hue is providing us an addition feature, that is sex. we have x,y for day, total_bill respectively. by adding hue we have catagorised data accoring to sex.

## violin plot
it is caombination of boxplot and a KDE plot

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=True)
```

## strip plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex',jitter=True,dodge=True)
```

#### jitter will spread out data, so they don't stack over each other.
#### dodge- will seperate the two categories(i.e., male, female) 

## swarm plot 
combination of violin and strip plot

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex')
sns.swarmplot(x='day',y='total_bill',data=tips_df,color='white',hue='sex')
```

# matrix plots

## Heatmaps

correlation matrix gives the matrix with columns(features) of data frame to the columns(features) of the data frame.

this tells us how influencial a feature to the final output.

#### note: here the out put refers to each feture for each row. 
#### in other words we can say by seeing correlation matrix we can say how much dependent a feature w.r.t another feature. 

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)

crash_mx=crash_df.corr() #this gives us the correlation matrix.
crash_mx
```

#### by seeing the diagonal, it is obvious that all are ones, since a feture related to it self by 100%

```python
sns.heatmap(crash_mx,annot=True, cmap='Blues') #annot-means annotation to add values at the middle
```

## pair grid

```python
iris=sns.load_dataset('iris')
iris_g=sns.PairGrid(iris,hue='species')
iris_g.map_diag(plt.hist)  #we want histogram along the diagonal
iris_g.map_offdiag(plt.scatter) #we want scatter along off-diag
#iris_g.map_lower(plt.scatter)
#iris_g.map_upper(plt.kdeplot)
```

### we can do custom plots as 

```python
iris_g=sns.PairGrid(data=iris, x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'],hue='species')
iris_g.map(plt.scatter)
```
